## 1 Function to generate random numbers without using any Python libraries

```python
#Function to update the seed value each time a random number is generated
def random(m=2**32, a=1103515245, c=12345): 
    random.new = (a*random.new + c) % m 
    return random.new/m 

#Function that develps n random numbers and appends them to the list 'l'
def rng(n=100, seed=1):
    global l
    global n_iter
    random.new = seed
    n_iter = n
    l = []
    for i in range(n_iter):
        s= random()
        l.append(s)
    return l
print(rng ())
```
